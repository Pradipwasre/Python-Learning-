# Day 5: Operators in Action – The Fundamentals of Python Expressions

Operators are the building blocks of Python expressions. They allow us to perform calculations, make decisions, and update values. Let’s break them down with simple examples and tips.

---

## 1. Arithmetic Operators
Used for mathematical operations.

```python
a = 10
b = 3
```

```python
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

Tips to Understand:

- / always gives a float (decimal).
- // removes the decimal part (floor division).
- % is remainder — great for checking divisibility.
- ** is power — useful for squares, cubes, etc.


# 2. Logical Operators

- Work with boolean values (True or False).

```python
print("True and True:", True and True)
print("True and False:", True and False)
print("True or False:", True or False)
print("False or False:", False or False)
print("not True:", not True)
print("not False:", not False)
```

Tips to Understand:

- and → both must be True.
- or → at least one must be True.
- not → flips the value.

**Think of real life:**
- "Can vote" = age ≥ 18 and has ID.
- "Can relax" = weekend or holiday.
- "Can play outside" = not raining.

# 3. Assignment Operators

- Used to assign and update values.

```python
x = 10
print("x =", x)

x += 5
print("After +=:", x)

x -= 3
print("After -=:", x)

x *= 2
print("After *=:", x)

x /= 4
print("After /=:", x)
```

# 4. Comparison Operators

Used to compare values. Always return True or False.

```python
print("10 == 10:", 10 == 10)
print("10 != 5:", 10 != 5)
print("10 > 5:", 10 > 5)
print("5 < 10:", 5 < 10)
print("10 >= 10:", 10 >= 10)
print("5 <= 10:", 5 <= 10)
```

Tips to Understand:

- " == checks equality.
- " != checks inequality.
- " > and < check greater/less.
- " >= and <= include equality.
- " Useful for conditions like marks, age, prices, etc.

Best Ways to Learn Operators

- Experiment: Change values and see results.
- Relate to real life: Discounts, averages, comparisons.
- Practice small problems: Sum, average, remainder, comparisons.
- Remember outputs: Logical and comparison operators always give True or False.
- Keep it simple: Use clear variable names (marks, price, age) to avoid confusion.

## Do’s and Don’ts

###Do’s

- Use operators in small, practical examples.
- Print results to see outputs clearly.
- Practice with numbers and booleans.
- Use assignment operators to simplify updates.

### Don’ts

- Don’t confuse = (assignment) with == (comparison).
- Don’t forget / vs // difference.
- Don’t use vague variable names.
- Don’t skip practice , repetition builds confidence.

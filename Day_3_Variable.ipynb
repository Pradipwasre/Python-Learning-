3. Data Types
Python has different data types such as:

- String (str): Text inside quotes
- Integer (int): Whole numbers
- Float (float): Decimal numbers
- Boolean (bool): True or False values
- Complex (complex): Numbers with real and imaginary parts

````python
print(type("Hello world"))
print(type(45))
print(type(10.5))
print(type(True))
print(type(3 + 4j))
````

## 6. Rules for Variable Names

Correct:

- Must start with a letter or underscore
- Can contain letters, numbers, underscore
- Case-sensitive (Name and name are different)
- Use descriptive names

Incorrect:

- Cannot start with a number (1student)
- Cannot use special characters (student-name)
- Cannot use Python keywords (print, for, class)
- Cannot have spaces (student name)

## 7. Boolean Values

- Boolean represents True or False.

````python
is_raining = False
is_day = True
````

````python
print(type(is_raining))
print(type(is_day))
````

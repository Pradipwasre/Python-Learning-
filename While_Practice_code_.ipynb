```python
# Print numbers from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# 2. Print even numbers from 2 to 10
i = 2
while i <= 10:
    print(i)
    i += 2
```

```python
# 3. Countdown from 5 to 1
i = 5
while i > 0:
    print(i)
    i -= 1
```

```python
# 4. Sum of first 5 numbers

i = 1
total = 0

while i <= 5:
    total += i
    i += 1

print("Total =", total)
```

```python
# 5. Print each character of a string
text = "hello"
i = 0

while i < len(text):
    print(text[i])
    i += 1
```

```python
# 6. Ask user until they type 'exit'
word = ""

while word != "exit":
    word = input("Type something (type 'exit' to stop): ")
```

```python
# 7. Print elements of a list
marks = [10, 20, 30, 40]
i = 0

while i < len(marks):
    print(marks[i])
    i += 1
```

```python
# 8. Multiply number until it reaches 100
num = 2

while num < 100:
    print(num)
    num *= 2
```

```python
# 9. Print table of 3
i = 1

while i <= 10:
    print(3, "x", i, "=", 3 * i)
    i += 1
```

```python
# 10. Keep adding numbers until total reaches 50
total = 0
i = 1

while total < 50:
    total += i
    print("Added", i, "Total =", total)
    i += 1
```

```python
correct_mark = 94
guess = 0

while guess != correct_mark:
    guess = int(input("Guess the mark (between 0-100): "))

    if guess < correct_mark:
        print("Too low! Try again.")
    elif guess > correct_mark:
        print("Too high! Try again.")
    else:
        print(f"Correct! The mark is {correct_mark}!")
```

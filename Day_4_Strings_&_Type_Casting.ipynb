# 1. Type Casting:

**Type casting means converting one data type into another.**

- int() → converts to integer
- float() → converts to float
- str() → converts to string

```python
num = 10 # it is in the Int data type
```

```python
# float() → converts to float
print(float(num))
```

```python
value = "25"
```

```python
# int() : converts to integer
print(int(value))
```

```python
pi = 3.14
```

```python
# str() → converts to string
print(str(pi))
```

# 2. String Indexing

***Strings are sequences of characters. Each character has an index (position).***

- Index starts at 0 (first character).
- Negative index -1 means last character.



```python
text = "Python"
```

```python
print(text[0])   # P (first character)
print(text[2])   # t (third character)
print(text[-1])  # n (last character)
```

# 4. String Slicing

- We can extract parts of a string using slicing `[start:end]`.

```python
a = "Hey there"
```

```python
print(a[:3])
print(a[0:3])
print(a[4:9])
print(a[-5:])   # End is empty
print(a[4:])    # End is empty
```

*now you try to chnage the number and check for the ouput*

- [start:end]
- If start is empty → begins from start
- If end is empty → goes till end

```python
a = "Hello there, today we are learning something new"
```

```python
print(a[6:])
```

```python
print(a[-3:])
```

# Python Strings & Type Casting – Summary Notes

## Type Casting
- Convert between data types:
  - `int()` → integer
  - `float()` → float
  - `str()` → string

## String Indexing
- Index starts at 0, negative index counts from end.
- Example:
  - `"Python"[0]` → P
  - `"Python"[2]` → t
  - `"Python"[-1]` → n

## String Slicing
- Extract substring with `[start:end]`.
- Examples:
  - `"Hey there"[0:3]` → Hey
  - `"Hey there"[4:9]` → there
  - `"Hey there"[-5:]` → there
  - `"Hey there"[4:]` → there
  - `"Hello there, today we are learning something new"[6:]` → there, today we are learning something new
  - `"Hello there, today we are learning something new"[-3:]` → new

## Extracting from Email
- `"student@gmail.com"[0:7]` → student

## String Methods
- `.upper()` → converts to uppercase  
- `.lower()` → converts to lowercase  
- `.title()` → converts to title case  

Examples:
- `"hello world".upper()` → HELLO WORLD  
- `"HEY THERE".lower()` → hey there  
- `"HEY THERE all, hows goin".title()` → Hey There All, Hows Goin  

## Searching
- `.find()` returns index or -1 if not found.
- `"Python is easy to learn".find("easy")` → 10  
- `"Python is easy to learn".find("a")` → 11  
- `"Python is easy to learn".find("china")` → -1  

## Replace
- `"Python is easy to learn".replace("easy","Best")` → Python is Best to learn  
- `"Python is Best to learn".replace("Best","Amazing")` → Python is Amazing to learn  

## Split & Join
- `.split()` → breaks string into list of words  
- `" ".join(list)` → joins list back into string  

Example:
- `"Python is Amazing to learn".split()` → ['Python','is','Amazing','to','learn']  
- `" ".join(['Python','is','Amazing','to','learn'])` → Python is Amazing to learn  

## Concatenation
- `"Amit" + " " + "kumar"` → Amit kumar  

## Subset of Words
- `"hello world how are you".split()[0:2]` → ['hello','world']  
- `" ".join(['hello','world'])` → hello world  

---

## Key Points
- **Type Casting** → `int`, `float`, `str`  
- **Indexing** → single characters  
- **Slicing** → parts of string  
- **Methods** → `upper()`, `lower()`, `title()`, `find()`, `replace()`  
- **Split & Join** → break/join words  
- **Concatenation** → combine strings
